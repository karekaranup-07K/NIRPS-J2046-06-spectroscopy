````python
# nirps_analysis.py
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from scipy.ndimage import gaussian_filter1d
from scipy.signal import find_peaks
from scipy.optimize import curve_fit
from scipy.interpolate import interp1d
from astropy.table import Table
import warnings
warnings.filterwarnings("ignore", category=RuntimeWarning)
````

````python
def load_spectrum(filename, flux_choice='FLUX_TELL_CAL'):
    """
    Load the SPECTRUM binary table from ESO Phase 3 file.
    flux_choice: 'FLUX', 'FLUX_CAL', 'FLUX_TELL_CAL'
    Returns: wave_nm (1D), flux (1D), err (1D), ancillary dict (SNR, ATM_TRANSM, QUAL arrays)
    """
    with fits.open(filename) as hdul:
        data = hdul[1].data
        cols = hdul[1].columns
        # WAVE units are Å in this file; convert to nm
        wave = data['WAVE'][0] / 10.0  # nm
        flux = data[flux_choice][0]
        # Choose matching error column name mapping
        err_map = {
            'FLUX': 'ERR',
            'FLUX_CAL': 'ERR_CAL',
            'FLUX_TELL_CAL': 'ERR_TELL_CAL'
        }
        errcol = err_map.get(flux_choice, 'ERR')
        err = data[errcol][0] if errcol in data.names else np.full_like(flux, np.nan)

        ancillary = {
            'SNR': data['SNR'][0] if 'SNR' in data.names else None,
            'ATM_TRANSM': data['ATM_TRANSM'][0] if 'ATM_TRANSM' in data.names else None,
            'QUAL': data['QUAL'][0] if 'QUAL' in data.names else None
        }
    return wave, flux, err, ancillary
````

````python
def build_mask(wave, flux, err, ancillary, atm_threshold=0.6, qual_bad_values=[1,2,4,8]):
    """
    Create boolean mask for valid data points.
    - Remove NaNs, negative/zero err
    - Mask telluric-dominated regions by ATM_TRANSM < threshold
    - Mask flagged QUAL bits if provided (QUAL is integer flags array)
    """
    mask = np.isfinite(flux) & np.isfinite(wave)
    if err is not None:
        mask &= np.isfinite(err)
        # optional: require err > 0
        mask &= (err > 0)
    # mask by atm transmission
    at = ancillary.get('ATM_TRANSM', None)
    if at is not None:
        mask &= (at > atm_threshold)

    # mask by QUAL if present (QUAL integer mask per pixel)
    qual = ancillary.get('QUAL', None)
    if qual is not None:
        # Keep where QUAL == 0 (no problems). If qual_bad_values are bit flags, adapt as needed.
        try:
            mask &= (qual == 0)
        except:
            pass
    return mask
````

````python
def rebin_linear(wave, flux, err=None, dv=2.0):
    """
    Rebin to a linear wavelength grid in velocity step dv (km/s).
    Returns new_wave, new_flux, new_err.
    """
    c = 299792.458  # km/s
    # create log-lambda grid -> constant velocity
    lam0 = wave[0]
    lamN = wave[-1]
    nl = int(np.floor(np.log(lamN/lam0) / np.log(1 + dv/ c))) + 1
    new_wave = lam0 * (1 + dv/c) ** np.arange(nl)
    interp_flux = interp1d(wave, flux, bounds_error=False, fill_value=np.nan)
    new_flux = interp_flux(new_wave)
    if err is not None:
        interp_err = interp1d(wave, err, bounds_error=False, fill_value=np.nan)
        new_err = interp_err(new_wave)
    else:
        new_err = None
    return new_wave, new_flux, new_err
````

````python
def estimate_continuum(wave, flux, mask=None, window_nm=30.0, poly_order=3):
    """
    Estimate continuum using a rolling median + low-order polynomial fit
    window_nm: smoothing window in nm
    Returns continuum array and normalized flux.
    """
    if mask is None:
        mask = np.isfinite(flux)
    # coarse smoothing via gaussian filter on flux (requires roughly uniform sampling)
    # approximate smoothing sigma in pixels
    # approximate pixel scale:
    pix_scale = np.median(np.diff(wave))
    sigma_pix = max(1.0, int(window_nm / pix_scale / 2.355))
    smooth = gaussian_filter1d(np.nan_to_num(flux), sigma=sigma_pix)
    # fit low-order polynomial to smooth curve
    # mask strong lines via iterative sigma-clipping on residuals
    good = mask.copy()
    for _ in range(4):
        x = wave[good]
        y = smooth[good]
        coeffs = np.polyfit(x, y, poly_order)
        fit = np.polyval(coeffs, wave)
        resid = smooth - fit
        std = np.nanstd(resid[good])
        good = good & (np.abs(resid) < 3*std)
    continuum = np.polyval(coeffs, wave)
    norm_flux = flux / continuum
    return continuum, norm_flux
````

````python
def gaussian(x, a, mu, sigma, bg):
    return a * np.exp(-0.5*((x-mu)/sigma)**2) + bg

def fit_line(wave_region, flux_region, err_region=None, p0=None, fit_voigt=False):
    """
    Fit a Gaussian to a line (works for absorption or emission).
    Returns fit parameters + covariance.
    """
    # p0: [amplitude, center, sigma, background]
    if p0 is None:
        idx = np.argmax(np.abs(flux_region - np.median(flux_region)))
        mu0 = wave_region[idx]
        a0 = (flux_region[idx] - np.median(flux_region))
        sigma0 = 0.2  # nm initial
        bg0 = np.median(flux_region)
        p0 = [a0, mu0, sigma0, bg0]
    try:
        popt, pcov = curve_fit(gaussian, wave_region, flux_region, p0=p0, sigma=err_region, absolute_sigma=False, maxfev=10000)
    except Exception as e:
        popt, pcov = None, None
    return popt, pcov
````

````python
def measure_ew(wave_region, norm_flux_region):
    """
    Simple EW (nm) integration for normalized spectrum around a line:
      EW = sum((1 - F) * dw)  for absorption (if normalized continuum ~1)
    Works for both emission (negative EW) and absorption (positive EW)
    Requires normalized flux (continuum = 1).
    """
    dw = np.diff(wave_region)
    # Make dw same length as wave array: append last dw value
    dw = np.concatenate([dw, [dw[-1]]])
    ew = np.nansum((1.0 - norm_flux_region) * dw)
    return ew
````

````python
def cross_correlation_rv(wave, norm_flux, mask_lines=None, vmin=-300, vmax=300, dv=0.5):
    """
    Compute cross-correlation function (CCF) by shifting template = mask_lines,
    or by self-CCF (autocorrelation).
    If mask_lines is None, use a simple line mask built from strong peaks.
    Returns velocities array and CCF values; RV at CCF peak.
    """
    c = 299792.458  # km/s
    # Build a mask/template: delta functions at strong line positions
    if mask_lines is None:
        # detect strong absorption features in normalized spectrum (peaks in 1 - flux)
        inv = 1.0 - norm_flux
        peaks, _ = find_peaks(inv, height=np.nanpercentile(inv, 99)*0.3, distance=5)
        mask_lines = wave[peaks]
        if len(mask_lines) < 5:
            # broaden detection threshold
            peaks, _ = find_peaks(inv, height=np.nanpercentile(inv, 95)*0.2, distance=3)
            mask_lines = wave[peaks]

    vel_grid = np.arange(vmin, vmax+dv, dv)
    ccf = np.zeros_like(vel_grid, dtype=float)
    # For each velocity, doppler-shift mask and compute dot product with spectrum
    for i, v in enumerate(vel_grid):
        shifted = mask_lines * (1 + v/c)
        # make a template spectrum by putting delta spikes at shifted positions
        template = np.zeros_like(wave)
        interp_template = interp1d(wave, template, bounds_error=False, fill_value=0.0)
        # Instead, build template by summing gaussians at shifted line centers
        for l in shifted:
            idx = np.searchsorted(wave, l)
            if 0 <= idx < len(wave):
                template[max(0, idx-2):min(len(wave), idx+3)] += 1.0
        # compute correlation
        ccf[i] = np.nansum((template - np.nanmean(template)) * (norm_flux - np.nanmean(norm_flux)))
    rv = vel_grid[np.argmax(ccf)]
    return vel_grid, ccf, rv
````

````python
def plot_line_region(wave, flux, err, wcenter, width=2.0, continuum=None, show_fit=True):
    mask = (wave >= wcenter - width) & (wave <= wcenter + width)
    x = wave[mask]; y = flux[mask]; e = err[mask] if err is not None else None
    plt.figure(figsize=(7,4))
    plt.plot(x, y, '-', lw=0.7, label='flux')
    if e is not None:
        plt.fill_between(x, y-e, y+e, alpha=0.3, label='err')
    if continuum is not None:
        plt.plot(x, continuum[mask], '--', label='continuum')
        plt.plot(x, y/continuum[mask], ':', label='normed')
    plt.xlabel('Wavelength [nm]')
    plt.title(f'Line around {wcenter:.2f} nm')
    plt.legend()
    plt.tight_layout()
    plt.show()
````

````python
# 0) parameters
filename = "ADP.2025-07-31T08_21_26.491.fits"
flux_choice = 'FLUX_TELL_CAL'  # recommended

# 1) Load
wave, flux, err, ancillary = load_spectrum(filename, flux_choice=flux_choice)
print(f"Loaded spectrum: {len(wave)} points. Wavelength range: {wave[0]:.1f}-{wave[-1]:.1f} nm")

# 2) Mask bad / telluric
mask = build_mask(wave, flux, err, ancillary, atm_threshold=0.65)
print(f"Valid points after mask: {np.count_nonzero(mask)}")

# 3) Optionally rebin to constant velocity grid (for CCF)
# new_wave, new_flux, new_err = rebin_linear(wave[mask], flux[mask], err[mask], dv=1.0)
# Or keep original sampling for continuum fits:
w_sub = wave[mask]; f_sub = flux[mask]; e_sub = err[mask] if err is not None else None

# 4) Continuum estimate and normalize
continuum, norm_flux = estimate_continuum(w_sub, f_sub, mask=np.isfinite(f_sub), window_nm=30.0, poly_order=3)
print("Estimated continuum. Median continuum:", np.nanmedian(continuum))

# 5) Plot a few regions interactively (examples)
plot_line_region(w_sub, f_sub, e_sub, 1080.0, width=2.5, continuum=continuum)
plot_line_region(w_sub, f_sub, e_sub, 1096.0, width=2.5, continuum=continuum)

# 6) Detect strong features (example in normalized spectrum)
# Convert to full array aligned with w_sub
peaks, props = find_peaks(1.0 - norm_flux, height=np.nanpercentile(1.0-norm_flux, 97)*0.2, distance=3)
line_centers = w_sub[peaks]
print("Detected candidate line centers (nm):", np.round(line_centers, 3))

# 7) Fit each strong line and measure EW
results = []
for c in line_centers:
    m = (w_sub >= c-1.0) & (w_sub <= c+1.0)
    if np.count_nonzero(m) < 10:
        continue
    popt, pcov = fit_line(w_sub[m], f_sub[m], err_region=e_sub[m], p0=None)
    ew = measure_ew(w_sub[m], (f_sub[m] / continuum[m]))
    results.append({'center_guess': c, 'fit_params': popt, 'ew_nm': ew})
    
# 8) Cross-correlation RV (simple)
vels, ccf, rv = cross_correlation_rv(w_sub, norm_flux, mask_lines=None, vmin=-200, vmax=200, dv=0.5)
print("Simple CCF RV estimate (km/s):", rv)

# 9) Plot CCF
plt.figure(figsize=(8,3))
plt.plot(vels, ccf)
plt.axvline(rv, color='red', linestyle='--', label=f'RV={rv:.2f} km/s')
plt.xlabel('Velocity [km/s]')
plt.ylabel('CCF')
plt.legend()
plt.tight_layout()
plt.show()

# 10) Save results table
from astropy.table import Table
tbl = Table(rows=[(r['center_guess'], r['fit_params'][1] if r['fit_params'] is not None else np.nan, r['ew_nm'])
                  for r in results],
            names=('center_guess_nm', 'fit_center_nm', 'EW_nm'))
tbl.write('line_measurements.ecsv', overwrite=True)
print("Saved line_measurements.ecsv")
````

````python
# nirps_analysis.py (v2.2: Fixed empty peaks_abs error)
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from scipy.ndimage import gaussian_filter1d
from scipy.signal import find_peaks
from scipy.optimize import curve_fit
from astropy.table import Table
import warnings
warnings.filterwarnings("ignore", category=RuntimeWarning)

# a) Load spectrum
def load_spectrum(filename, flux_choice='FLUX_TELL_CAL'):
    """Load with confirmed columns. flux_choice: 'FLUX', 'FLUX_CAL', 'FLUX_TELL_CAL'"""
    with fits.open(filename) as hdul:
        data = hdul[1].data
        wave = data['WAVE'][0] / 10.0  # nm
        flux = data[flux_choice][0]
        err_map = {'FLUX': 'ERR', 'FLUX_CAL': 'ERR_CAL', 'FLUX_TELL_CAL': 'ERR_TELL_CAL'}
        err = data[err_map[flux_choice]][0]
        ancillary = {
            'SNR': data['SNR'][0],
            'ATM_TRANSM': data['ATM_TRANSM'][0],
            'QUAL': data['QUAL_TELL_CAL'][0] if flux_choice == 'FLUX_TELL_CAL' else data['QUAL'][0]
        }
    return wave, flux, err, ancillary

# b) Quality mask & telluric masking
def build_mask(wave, flux, err, ancillary, atm_threshold=0.65, min_snr=3):
    """Enhanced mask with SNR cut."""
    mask = np.isfinite(flux) & np.isfinite(err) & (err > 0)
    mask &= (ancillary['ATM_TRANSM'] > atm_threshold)
    mask &= (ancillary['QUAL'] == 0)
    mask &= (ancillary['SNR'] > min_snr)
    return mask

# c) Continuum estimation & normalization
def estimate_continuum(wave, flux, mask=None, window_nm=30.0, poly_order=3):
    """Estimate continuum using a rolling median + low-order polynomial fit."""
    if mask is None:
        mask = np.isfinite(flux)
    sigma_pix = max(1.0, int(window_nm / np.median(np.diff(wave)) / 2.355))
    smooth = gaussian_filter1d(np.nan_to_num(flux), sigma=sigma_pix)
    good = mask.copy()
    for _ in range(4):
        x = wave[good]
        y = smooth[good]
        coeffs = np.polyfit(x, y, poly_order)
        fit = np.polyval(coeffs, wave)
        resid = smooth - fit
        std = np.nanstd(resid[good])
        good = good & (np.abs(resid) < 3*std)
    continuum = np.polyval(coeffs, wave)
    norm_flux = flux / continuum
    return continuum, norm_flux

# d) Peak detection and Gaussian/Voigt fitting
def gaussian(x, a, mu, sigma, bg):
    return a * np.exp(-0.5*((x-mu)/sigma)**2) + bg

def fit_line(wave_region, flux_region, err_region=None, p0=None):
    """Fit a Gaussian to a line."""
    if p0 is None:
        idx = np.argmax(np.abs(flux_region - np.median(flux_region)))
        mu0 = wave_region[idx]
        a0 = (flux_region[idx] - np.median(flux_region))
        sigma0 = 0.2
        bg0 = np.median(flux_region)
        p0 = [a0, mu0, sigma0, bg0]
    try:
        popt, pcov = curve_fit(gaussian, wave_region, flux_region, p0=p0, sigma=err_region, absolute_sigma=False, maxfev=10000)
    except Exception:
        popt, pcov = None, None
    return popt, pcov

def measure_ew(wave_region, norm_flux_region):
    """Measure equivalent width (nm) for normalized spectrum."""
    dw = np.diff(wave_region)
    dw = np.concatenate([dw, [dw[-1]]])
    ew = np.nansum((1.0 - norm_flux_region) * dw)
    return ew

# e) Quasar redshift refinement via CCF
def cross_correlation_quasar_z(wave, norm_flux, z_guess=0.1287, dv=50.0):
    """CCF for quasar systemic redshift using NIR emission lines."""
    template_lines_rest = [1094, 1282, 1005, 1379]  # Paγ, Paβ, He I, C IV (rest nm)
    c = 299792.458
    z_min, z_max = z_guess - 0.02, z_guess + 0.02
    z_grid = np.arange(z_min, z_max, dv / c)
    ccf = np.zeros(len(z_grid))
    em_peaks, _ = find_peaks(norm_flux, height=1.05, distance=20)
    em_mask = np.zeros_like(norm_flux); em_mask[em_peaks] = 1.0
    for i, z in enumerate(z_grid):
        shifted = np.array(template_lines_rest) * (1 + z)
        temp_mask = np.zeros_like(wave)
        for lam in shifted:
            if np.any((wave > lam - 5) & (wave < lam + 5)):
                idx = np.argmin(np.abs(wave - lam))
                temp_mask[max(0, idx-3):min(len(wave), idx+3)] = 1.0
        temp_norm = (temp_mask - np.nanmean(temp_mask)) / (np.nanstd(temp_mask) + 1e-10)
        em_norm = (em_mask - np.nanmean(em_mask)) / (np.nanstd(em_mask) + 1e-10)
        ccf[i] = np.nansum(temp_norm * em_norm * np.isfinite(norm_flux))
    best_z = z_grid[np.argmax(ccf)]
    return z_grid, ccf, best_z

# Workflow execution
if __name__ == "__main__":
    # Parameters
    filename = "ADP.2025-07-31T08_21_26.491.fits"
    flux_choice = 'FLUX_TELL_CAL'
    z_lit = 0.1287

    # 1. Load
    wave, flux, err, ancillary = load_spectrum(filename, flux_choice)
    print(f"Loaded: {len(wave)} pts, {wave[0]:.0f}-{wave[-1]:.0f} nm, median flux {np.nanmedian(flux):.2e} erg/s/cm²/Å")

    # 2. Mask
    mask = build_mask(wave, flux, err, ancillary)
    w_sub, f_sub, e_sub = wave[mask], flux[mask], err[mask]
    print(f"Masked: {len(w_sub)} pts ({100*len(w_sub)/len(wave):.1f}%)")

    # 3. Continuum normalize
    continuum, norm_flux = estimate_continuum(w_sub, f_sub, window_nm=30.0, poly_order=3)
    norm_err = e_sub / continuum

    # 4. Plot full normalized spectrum
    plt.figure(figsize=(12, 5))
    plt.plot(wave, flux / np.nanmedian(flux), 'k-', lw=0.5, label='Raw Norm. Flux')
    plt.fill_between(wave, (flux-err)/np.nanmedian(flux), (flux+err)/np.nanmedian(flux), alpha=0.3, color='gray')
    plt.plot(w_sub, norm_flux, 'b-', lw=0.8, alpha=0.7, label='TELL_CAL + Mask + Continuum')
    plt.ylim(0.5, 1.5)
    plt.xlabel('Wavelength [nm]'); plt.ylabel('Normalized Flux'); plt.title(f'NIRPS Spectrum of J2046+06 (z_lit={z_lit})')
    plt.legend(); plt.tight_layout(); plt.show()

    # 5. Detect strong features
    inv_flux = 1.0 - norm_flux
    peaks_abs, props_abs = find_peaks(inv_flux, height=np.nanmean(inv_flux) + 3*np.nanstd(inv_flux), distance=20, prominence=0.1, rel_height=0.5)
    em_peaks, _ = find_peaks(norm_flux, height=1.1, distance=20)
    abs_centers = w_sub[peaks_abs[:15]]
    em_centers = w_sub[em_peaks[:5]]
    line_centers = np.sort(np.concatenate([abs_centers, em_centers]))
    print(f"Top line centers (nm): {np.round(line_centers[:10], 2)}")

    # 6. Fit & EW for top 10 (fixed type: robust peak check)
    results = []
    for i, c in enumerate(line_centers[:10]):
        m = (w_sub >= c-2.0) & (w_sub <= c+2.0)
        if np.sum(m) < 30: continue
        popt, pcov = fit_line(w_sub[m], f_sub[m], e_sub[m])
        ew = measure_ew(w_sub[m], norm_flux[m])
        diffs = np.diff(w_sub[m])
        dw_err = np.append(diffs**2, [diffs[-1]**2])
        ew_err = np.sqrt(np.sum((norm_err[m]**2) * dw_err))
        # Robust type assignment: check nearest absorption peak
        if len(peaks_abs) > 0:
            abs_idx = np.argmin(np.abs(w_sub[peaks_abs] - c))
            peak_wl = w_sub[peaks_abs[abs_idx]]
            line_type = 'abs' if abs(peak_wl - c) < 2.0 and abs(inv_flux[peaks_abs[abs_idx]] - (1 - np.mean(norm_flux[m]))) < 0.05 else 'em'
        else:
            line_type = 'em'
        fit_center = popt[1] if popt is not None else np.nan
        results.append({'center': c, 'type': line_type, 'ew_nm': ew, 'ew_err': ew_err, 'fit_center': fit_center})

    # Plot top line with residuals
    if results:
        top_c = results[0]['center']
        m_top = (w_sub >= top_c-2.0) & (w_sub <= top_c+2.0)
        x_top, y_top = w_sub[m_top], norm_flux[m_top]
        popt_top, _ = fit_line(x_top, y_top * continuum[m_top], None)
        if popt_top is not None:
            fit_y = gaussian(x_top, *popt_top)
            resid = y_top - (fit_y / np.mean(continuum[m_top]))
        else:
            resid = np.zeros_like(y_top)
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(6,6), sharex=True)
        ax1.plot(x_top, y_top, 'b-', label='Norm Flux')
        if popt_top is not None:
            ax1.plot(x_top, fit_y / np.mean(continuum[m_top]), 'r--', label='Gaussian Fit')
        ax1.axvline(top_c, color='k', ls='--', label=f'Center {top_c:.1f} nm')
        ax1.set_ylabel('Norm Flux'); ax1.legend(); ax1.set_title(f'Top Line at {top_c:.1f} nm (EW={results[0]["ew_nm"]:.3f} nm)')
        ax2.plot(x_top, resid, 'g-', label='Residuals')
        ax2.axhline(0, color='k', ls=':'); ax2.set_xlabel('Wavelength [nm]'); ax2.set_ylabel('Residual'); ax2.legend()
        plt.tight_layout(); plt.show()

    # 7. Quasar z refiner
    z_grid, ccf_z, best_z = cross_correlation_quasar_z(w_sub, norm_flux, z_guess=z_lit)
    print(f"Refined z: {best_z:.4f} (literature {z_lit})")

    # Plot CCF
    plt.figure(figsize=(8,3)); plt.plot(z_grid*1000, ccf_z); plt.axvline(best_z*1000, color='r', ls='--', label=f'z={best_z:.4f}')
    plt.xlabel('Redshift (x1000)'); plt.ylabel('CCF'); plt.legend(); plt.show()

    # 8. Annotated table
    quasar_ids = {1005: 'He I (1083 Å rest)', 1094: 'Paγ (H I)', 1152: 'Lyα (H I)', 1282: 'Paβ (H I)',
                  1366: 'C IV (metal)', 1550: 'Si IV + C IV', 1875: 'Paα (H I)'}
    for r in results:
        obs_center = r['fit_center'] if not np.isnan(r['fit_center']) else r['center']
        matched = False
        for obs_nm, id_str in quasar_ids.items():
            if abs(obs_center - obs_nm) < 3.0:
                r['id'] = id_str
                matched = True
                break
        if not matched and r['ew_nm'] < 0.5 and r['type'] == 'abs':
            r['id'] = 'Lyα Forest (z_abs ~0.10)'
        else:
            r['id'] = 'Intrinsic/Unknown'

    tbl = Table(rows=[(r['center'], r['type'], r['ew_nm'], r['ew_err'], r['fit_center'], r['id'])
                     for r in results], names=('center_guess_nm', 'type', 'EW_nm', 'EW_err_nm', 'fit_center_nm', 'ID'))
    tbl.write('line_measurements_id.ecsv', overwrite=True)
    print(tbl)
    print("Saved line_measurements_id.ecsv")
````
